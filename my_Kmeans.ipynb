```
#Libraries
%matplotlib inline
import random
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
```

```
""" 
Created on Sat Feb 13 17:41:36 2021
@author: spanishkukli
@summary: Implement the Kmeans (clustering) algorithm with a random datapoint and obtain an output with a number K of clusters with their centroids.
"""
#Create random datapoints (2d)
v1 = []
v2 = []

V =int(input("n_V: "))
for i in range(V):
  v1.append(random.randint(-10,10)) 
  v2.append(random.randint(-10,10))

print(f" v1 -> {v1}\n v2 -> {v2}")
print("----" * V)

#Group datapoints
X = np.array(list(zip(v1, v2))).reshape(len(v1), 2) #[[x, y], [x, y]...]
#print(X)

#Plotting datapoints
plt.style.use("default")
plt.plot(X[:,0], X[:,1], "k.")
plt.axis([-11, 11, -11, 11])
plt.title('Kmeans')
plt.show()

#Ask for cluster nÂº
K = int(input("n_k: "))
#Train the model with datapoints and number of clusters 
k_means = KMeans(n_clusters=K).fit(X)
#Search for the centroid
centroid = k_means.cluster_centers_
#Plotting the centroids and changing the colors of the datapoints to differentiate them into groups
#Each color group will have a centroid
plt.scatter(X[:,0], X[:,1], s=10, c=k_means.labels_, cmap='rainbow')
plt.plot(centroid[:,0],centroid[:,1],'kx',markersize=5)
plt.title('Kmeans')
plt.show()
```
